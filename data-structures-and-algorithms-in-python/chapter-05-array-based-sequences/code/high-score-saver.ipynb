Q. We need to build a dynamic leaderboard (array-based), where each entry is stored as a tuple (username, score).
The array — called Leaderboard — must always remain sorted in descending order of scores.
Whenever a new score is added, the array should automatically find its correct position and maintain the order.

```python
class Leaderboard:
    """Dynamic sequence of high scores in descending order."""
    def __init__(self, initial_capacity=10):
        self._scores = [0]*initial_capacity
        self._n = 0     # number of actual scores

    def __grow(self):
        pass

    def add(self, score):
        if self._n < len(self._scores):
            j = self._n
            self._n += 1

        while j > 0 and self._scores[][j-1] < score:
            self._scores[j] = self._scores[j-1]
            j -= 1
        self._scores[j] = score
        
        pass

    def display(self):
        print("Current Leaderboard: {self._scores[:self._n]}")
```
